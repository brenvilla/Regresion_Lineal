## **Modelo matemático**

### **Dataframe DF_México**

**PRIVATE ROOM**

```python
private_mex_num = private_mex.select_dtypes(include='number')

#Encontramos las correlaciones entre las variables 
Corr_Factors_privatemex=private_mex_num.corr()
Corr_Factors_privatemex = Corr_Factors_privatemex.sort_values(by='number_of_reviews', ascending = False)
Corr_Factors_privatemex['number_of_reviews'].head()
```

```python
varcorr1 = Corr_Factors_privatemex.index[1]
varcorr2 = Corr_Factors_privatemex.index[2]
varcorr3 = Corr_Factors_privatemex.index[3]
print(varcorr1,varcorr2,varcorr3)
```

```python
#Declaramos las variables dependientes e independientes para la regresión lineal
Vars_Indep= private_mex[[varcorr1,varcorr2,varcorr3]]
Var_Dep= private_mex['number_of_reviews']
#Se define model como la función de regresión lineal
model= LinearRegression()
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
#Verificamos los coeficientes obtenidos para el modelo ajustado
model.__dict__
```

```python
#Predecimos los valores de 'number_of_reviews' a partir de la variable "number_of_reviews_ltm"
y_pred= model.predict(X=private_mex_num[[varcorr1,varcorr2,varcorr3]])
```

```python
#Insertamos la columna de predicciones en el DataFrame
private_mex_num.insert(0, 'Predicciones', y_pred)
private_mex_num.head()
```

```python
#Visualizamos la gráfica comparativa entre el valor real y el valor predicho

sns.scatterplot(x=varcorr1, y='number_of_reviews', color="blue", data=private_mex_num)
sns.scatterplot(x=varcorr1, y='Predicciones', color="red", data=private_mex_num)
```

```python
#Corroboramos cual es el coeficiente de Determinación de nuestro modelo
Deter_private_mex=model.score(X=Vars_Indep, y=Var_Dep)
print('Coeficiente de determinación (R2): %.2f' % Deter_private_mex)

#Corroboramos cual es el coeficiente de Correlación de nuestro modelo
Correl_private_mex=np.sqrt(Deter_private_mex)
print('Coeficiente de correlación: %.2f' % Correl_private_mex)
```
